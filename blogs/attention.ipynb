```python
# Scaled Dot-Product Attention
import torch
import torch.nn as nn
from torch.nn import functional as F
torch.manual_seed(42)

B, T, C = 2, 4, 2
x = torch.randn(B, T, C)
print(f"x[0]:\n {x[0]}")
query = nn.Linear(C, C)
key = nn.Linear(C, C)
value = nn.Linear(C, C)
tril = torch.tril(torch.ones(T, T))
q = query(x)
k = key(x)
v = value(x)
print(f"q[0]:\n {q[0]}")
print(f"v[0]:\n {v[0]}")

print("q @ k.transpose(1, 2)[0]")
print(q @ k.transpose(1, 2)[0])
attention_scores = (q @ k.transpose(1, 2)) / (k.shape[-1] ** 0.5)
print("attention_scores[0]")
print(attention_scores[0])
attention_scores = attention_scores.masked_fill(tril == 0, float("-inf"))
print("attention_scores[0]")
print(attention_scores[0])
atten_wei= F.softmax(attention_scores, dim=-1) 
print("atten_wei[0]")
print(atten_wei[0])
out = atten_wei @ v
print("out[0]")
print(out[0])
```

```python
# example for value weighted sum calculation
print(atten_wei[0,2:3,:]) 
print(v[0,:,:1])
print(atten_wei[0,2:3,:] @ v[0,:,:1])
```
